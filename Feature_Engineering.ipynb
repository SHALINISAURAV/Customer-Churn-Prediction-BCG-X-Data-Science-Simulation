```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
```

```python
# Load Dataset

df = pd.read_csv("clean_data_after_eda.csv")

print("Shape:", df.shape)
df.head()
```

```python
# Basic Info

df.info()
df.describe()
```

```python
# Remove Columns with Only 1 Unique Value

for col in df.columns:
    if df[col].nunique() == 1:
        print("Dropping:", col)
        df.drop(columns=col, inplace=True)

print("Shape after dropping constant columns:", df.shape)
```

```python
# Drop ID Columns (if present)

id_cols = ['customer_id', 'id']  # edit if needed

for col in id_cols:
    if col in df.columns:
        df.drop(columns=col, inplace=True)

df.head()
```

```python
# Date Feature Engineering

# CONVERT DATE COLUMN...
date_cols = df.select_dtypes(include='object').columns

for col in date_cols:
    try:
        df[col] = pd.to_datetime(df[col])
        print(f"{col} converted to datetime")
    except:
        pass
```

```python
# Extract Date Components

for col in df.select_dtypes(include='datetime64[ns]').columns:
    
    df[col + "_year"] = df[col].dt.year
    df[col + "_month"] = df[col].dt.month
    df[col + "_day"] = df[col].dt.day
    df[col + "_dayofweek"] = df[col].dt.dayofweek
    
    df.drop(columns=col, inplace=True)

df.head()
```

```python
# Create Behavioral Features

# Create Average-Based Features
if 'total_spent' in df.columns and 'number_of_orders' in df.columns:
    df['avg_order_value'] = df['total_spent'] / (df['number_of_orders'] + 1)
```

```python
# Create Usage Intensity

if 'login_count' in df.columns and 'tenure_days' in df.columns:
    df['login_per_day'] = df['login_count'] / (df['tenure_days'] + 1)
```

```python
# Encode Categorical Features

# one hot encoding
df = pd.get_dummies(df, drop_first=True)

df.head()
```

```python
# Handle Skewed Features

# Log Transform Highly Skewed Columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    if abs(df[col].skew()) > 1:
        df[col] = np.log1p(df[col])
```
