```python
# Data manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")
```

```python
# Load the Data

price_df = pd.read_csv('price_data (1).csv')
client_df = pd.read_csv('client_data (1).csv')

print("Price Data Shape:", price_df.shape)
print("Client Data Shape:", client_df.shape)
```

```python
# Preview the Data

print("Price Data Preview:")
display(price_df.head())

print("\nClient Data Preview:")
display(client_df.head())
```

```python
# Data Structure & Types

print("Price Data Info:")
price_df.info()

print("\nClient Data Info:")
client_df.info()
```

```python
# Check Missing Values

print("Missing Values in Price Data:")
print(price_df.isnull().sum())

print("\nMissing Values in Client Data:")
print(client_df.isnull().sum())
```

```python
# Descriptive Statistics – Numeric Columns

print("Price Data - Numeric Summary")
display(price_df.describe())

print("\nClient Data - Numeric Summary")
display(client_df.describe())
```

```python
# Descriptive Statistics – Categorical Columns

print("Price Data - Categorical Summary")
display(price_df.describe(include='object'))

print("\nClient Data - Categorical Summary")
display(client_df.describe(include='object'))
```

```python
# Unique Values per Column

print("Unique Values in Client Data:")
for col in client_df.columns:
    print(f"{col}: {client_df[col].nunique()}")
```

```python
# Convert Date Columns (If Needed)

# Example – change column name if needed
# price_df['price_date'] = pd.to_datetime(price_df['price_date'])
```

```python
# Distribution of Numeric Features

numeric_cols = client_df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(client_df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Outlier Detection (Boxplots)

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=client_df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# Categorical Variable Analysis

categorical_cols = client_df.select_dtypes(include='object').columns

for col in categorical_cols:
    if client_df[col].nunique() < 10:
        plt.figure(figsize=(6,4))
        sns.countplot(x=client_df[col])
        plt.xticks(rotation=45)
        plt.title(f"Countplot of {col}")
        plt.show()
```

```python
# Target Variable (If churn exists)

if 'churn' in client_df.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(x=client_df['churn'])
    plt.title("Churn Distribution")
    plt.show()
    
    print("Churn Percentage:")
    print(client_df['churn'].value_counts(normalize=True)*100)
```

```python
# Correlation Matrix

plt.figure(figsize=(10,8))
sns.heatmap(client_df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Merge Datasets (If Common Key Exists)

if 'client_id' in price_df.columns and 'client_id' in client_df.columns:
    merged_df = pd.merge(client_df, price_df, on='client_id', how='left')
    print("Merged Data Shape:", merged_df.shape)
    display(merged_df.head())
```
